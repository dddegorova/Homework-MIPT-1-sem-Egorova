```python
import math

n = int(input())
x = []
y = []
for i in range(n):
    xx, yy = map(int, input().split())
    x.append(xx)
    y.append(yy)


def can(d):
    used = [False] * n
    used[0] = True
    q = [0]
    while q:
        v = q.pop(0)
        for to in range(n):
            if not used[to]:
                dist = math.hypot(x[v] - x[to], y[v] - y[to])
                if dist <= d:
                    used[to] = True
                    q.append(to)
                    if to == 1:
                        return True
    return used[1]


lk = 0
r = 2000
for i in range(100):
    mid = (lk + r) / 2
    if can(mid):
        r = mid
    else:
        lk = mid
print("{:.3f}".format(r))
```

```python
c = int(input())
for _ in range(c):
    n, m = map(int, input().split())
    e = []
    for _ in range(m):
        x, y, t = map(int, input().split())
        e.append((x, y, t))
    d = [0] * n
    for i in range(n - 1):
        for x, y, t in e:
            if d[x] + t < d[y]:
                d[y] = d[x] + t
    pos = False
    for x, y, t in e:
        if d[x] + t < d[y]:
            pos = True
            break
    if pos:
        print("возможно")
    else:
        print("не возможно")
```
