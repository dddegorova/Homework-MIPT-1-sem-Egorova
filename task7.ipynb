```python
def pol(s):
    a = []
    n = ''
    for x in s:
        if x in '0123456789':
            n += x
        else:
            if n:
                a.append(n)
                n = ''
            if x != ' ':
                a.append(x)
    if n:
        a.append(n)
    p = {'+':1, '-':1, '*':2, '/':2}
    
    st = []
    obr = []
    
    for t in a:
        if t.isdigit():
            obr.append(t)
        elif t == '(':
            st.append(t)
        elif t == ')':
            while st and st[-1] != '(':
                obr.append(st.pop())
            st.pop()
        else:
            while st and st[-1] != '(' and p.get(st[-1],0) >= p.get(t,0):
                obr.append(st.pop())
            st.append(t)
    
    while st:
        obr.append(st.pop())
    
    st = []
    pr = []
    
    for t in reversed(a):
        if t.isdigit():
            pr.append(t)
        elif t == ')':
            st.append(t)
        elif t == '(':
            while st and st[-1] != ')':
                pr.append(st.pop())
            st.pop()
        else:
            while st and st[-1] != ')' and p.get(st[-1],0) > p.get(t,0):
                pr.append(st.pop())
            st.append(t)
    
    while st:
        pr.append(st.pop())
    
    return ' '.join(pr[::-1]), ' '.join(obr)

v = "(3+4*(2-1))/5"
pr, obr = pol(v)

print("Прямая:", pr)
print("Обратная:", obr)
```

## Упражнение 2

```python
def calc(s):
    a = s.split()
    st = []
    for x in a:
        if x in '+-*/':
            if len(st) < 2:
                return "Выражение составлено некорректно: мало чисел"
            b = st.pop()
            c = st.pop()
            if x == '+':
                st.append(c + b)
            elif x == '-':
                st.append(c - b)
            elif x == '*':
                st.append(c * b)
            elif x == '/':
                if b == 0:
                    return "Выражение составлено некорректно: деление на ноль"
                st.append(c / b)
        else:
            f = True
            d = 0
            for i in x:
                if i == '.':
                    d += 1
                elif i not in '0123456789':
                    f = False
                    break
            if f and d <= 1:
                if d == 0:
                    st.append(int(x))
                else:
                    st.append(float(x))
            else:
                return "Выражение составлено некорректно: не число"
    if len(st) != 1:
        return "Выражение составлено некорректно: много чисел"
    return st[0]


v = input()
print(calc(v))
```

```python
class Node:
    def __init__(self, data=None):
        self.data = data
        self.next = None
        self.prev = None


class List:
    def __init__(self):
        self.head = None
        self.tail = None

    def __len__(self):
        count = 0
        c = self.head

        while c:
            count += 1
            c = c.next
        return count

    def get(self, index):
        c = self.head
        i = 0
        while c and i <= index:
            if i == index:
                return c.data
            i += 1
            c = c.next
        return None

    def add(self, index, value):
        n = Node(value)
        if self.head is None:
            self.head = n
            self.tail = n
            return
        if index == 0:
            n.next = self.head
            self.head.prev = n
            self.head = n
            return
        c = self.head
        i = 0
        while c and i < index:
            if i == index - 1:
                break
            i += 1
            c = c.next
        if c:
            n.next = c.next
            n.prev = c
            if c.next:
                c.next.prev = n
            c.next = n
            if n.next is None:
                self.tail = n

    def remove(self, index):
        if self.head is None:
            return
        if index == 0:
            if self.head.next:
                self.head.next.prev = None
            self.head = self.head.next
            if self.head is None:
                self.tail = None
            return
        c = self.head
        i = 0
        while c and i < index:
            if i == index:
                break
            i += 1
            c = c.next
        if c:
            if c.prev:
                c.prev.next = c.next
            if c.next:
                c.next.prev = c.prev
            if c.next is None:
                self.tail = c.prev


l = List()
l.add(0, 10)
l.add(1, 20)
l.add(2, 30)
l.add(1, 15)
print("Длина:", len(l), "Элементы:", l.get(0), l.get(1), l.get(2), l.get(3))
```
